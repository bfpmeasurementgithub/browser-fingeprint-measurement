```python
company = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n']
attack_sum = []
normal_sum = []
for i in range(1,len(company)):
    attck_file = open('../' + company[i]+ '/' + company[i] +'_attack_bfp.csv','r',newline='')
    normal_file = open('../' + company[i]+ '/' + company[i] +'_normal_bfp.csv','r',newline='')
    attack_reader_lines = csv.reader(x.replace('\0', '') for x in attck_file)
    normal_reader_lines = csv.reader(x.replace('\0', '') for x in normal_file)
    sum_attack = 0
    sum_normal = 0
    for raw_data in attack_reader_lines:
        sum_attack += float(raw_data[1])  
    for raw_data in normal_reader_lines: 
        sum_normal += float(raw_data[1])  

    attack_sum.append(sum_attack)
    normal_sum.append(sum_normal)
```

```python
whole_attack = 0
whole_normal = 0
for i in attack_sum:
    whole_attack += i
for i in normal_sum:
    whole_normal += i
print(whole_attack, whole_normal, whole_attack + whole_normal, whole_attack/ (whole_attack + whole_normal)  )
```
