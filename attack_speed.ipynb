```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import csv
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from matplotlib import gridspec
from statsmodels import robust
import os
import csv
import pandas as pd
import sys
import time
from collections import defaultdict
from collections import Counter
```

```python
# The input data should follow this type
# in .csv file
# _timestamp,device_id,header,ip,ua,renderer,vendor,bfp


# This is the demo list, we remove the orignal data. 
pathFile = './'
fileList = os.listdir(pathFile)

# This is the demo list, We can not show the full list about that which also include the user information
botua = ["Requests", "bot", "Bot", "PhantomJS", "Python", "Java", 'python']
botvm = ['VMware','Unknown', 'Virtual']

for file in fileList:
    if 'csv' not in file:
        continue
    domain = os.path.abspath(pathFile)
    path = os.path.join(domain, file)
    with open(path, 'r', encoding='utf-8') as f:
        readData = csv.reader(f)
        result = list(readData)

        ip_dic = {}
        for raw in result:
            if raw[0] != '_timestamp' and len(raw) == 8:
                if len(raw[7]) == 1:
                        print (raw)
                if raw[3] not in ip_dic:
                    tem = set()
                    ip_dic[ raw[3] ] = [ [raw[0]], tem ]
                else:
                    ip_dic[ raw[3] ][0].append([raw[0]])
                if any(word in raw[4] for word in botua):
                        ip_dic[ raw[3] ][1].add('simple_bot')
                elif any(word in raw[5] for word in botvm):
                    ip_dic[ raw[3] ][1].add('vm')
                else:
                    bfp = raw[7].split('-')
                    if len(bfp) != 7:
                        ip_dic[ raw[3] ][1].add('bro_bot')
                    else:
                        bfp = bfp[1:]
                        if all(len(word) != 8 for word in bfp):
                            ip_dic[ raw[3] ][1].add('bro_bot')
                        else:
                            ip_dic[ raw[3] ][1].add('browser')
ip_dic = dict( sorted( ip_dic.items(), key = lambda i: len(i[1][0]), reverse = 1     )   )
simple_bot = []
vm = []
bro_bot = []
browser = []
for ele in ip_dic:
    # min_val is depends on the data, it can be changed in the different company data
    min_val = 100
    if len(ip_dic[ele][0]) > min_val:
        date_dic = defaultdict(list)
        result_arr = []
        for time in ip_dic[ele][0]:
            time_s = ''.join(time)
            date = time_s.split('T')[0]
            second_s = time_s.split('T')[1].split(':')
            second = float(second_s[0]) * 60 * 60 + float(second_s[1]) * 60 + float(second_s[2][:-1])
            date_dic[date].append(second)
        for date in date_dic:
            if len(date_dic[date]) > 10:
                end = max(date_dic[date])
                start = min(date_dic[date])
                result = (end - start) / len(date_dic[date])
                if result > 100:
                    continue
                result_arr.append(result)
        if len(result_arr) > 0:
            if len(ip_dic[ele][1]) > 1:
                continue
            if "simple_bot" in ip_dic[ele][1]:
                simple_bot += result_arr
            if "vm" in ip_dic[ele][1]:
                vm += result_arr
            if "bro_bot" in ip_dic[ele][1]:
                bro_bot += result_arr
            if "browser" in ip_dic[ele][1]:
                browser += result_arr
print("simple_bot",simple_bot, np.mean(simple_bot), np.median(simple_bot))
print("vm",vm, np.mean(vm), np.median(vm))
print("bro_bot",bro_bot, np.mean(bro_bot), np.median(bro_bot))
print("browser",browser, np.mean(browser), np.median(browser))
            
print("===================================================================")
simple_bot = []
vm = []
bro_bot = []
browser = []
for ele in ip_dic:
    if len(ip_dic[ele][0]) > 100:
        date_dic = defaultdict(list)
        result_arr = []
        for time in ip_dic[ele][0]:
            time_s = ''.join(time)
            date = time_s.split('T')[0]
            second_s = time_s.split('T')[1].split(':')
            second = float(second_s[0]) * 60 * 60 + float(second_s[1]) * 60 + float(second_s[2][:-1])
            date_dic[date].append(second)
        for date in date_dic:
            if len(date_dic[date]) > 10:
                end = max(date_dic[date])
                start = min(date_dic[date])
                result = (end - start) / len(date_dic[date])
                if result > 100:
                    continue
                result_arr.append(result)
        if len(result_arr) > 0:
            if "simple_bot" in ip_dic[ele][1]:
                simple_bot += result_arr
            if "vm" in ip_dic[ele][1]:
                vm += result_arr
            if "bro_bot" in ip_dic[ele][1]:
                bro_bot += result_arr
            if "browser" in ip_dic[ele][1]:
                browser += result_arr
print("simple_bot",simple_bot, np.mean(simple_bot), np.median(simple_bot))
print("vm",vm, np.mean(vm), np.median(vm))
print("bro_bot",bro_bot, np.mean(bro_bot), np.median(bro_bot))
print("browser",browser, np.mean(browser), np.median(browser))    
```
