```python
import csv
import pandas as pd
import sys
import time
import numpy as np
```

```python
# analyze the asn part
'''
The file like this style

signal,customer,id_type,browser,filter,tel_isattack,start_ts,end_ts,
num_distinct_pid_not_null,num_ids,entropy,collision_ratio,num_values,
num_devices,division,num_device_with_multiple_values,ratio_device_multiple_values,num_distinct

'''
file = open('./demo_asn.csv','r',encoding='utf-8',newline='')
reader_lines = csv.reader(file)
label = ['total', 'attack', 'normal']
no_none = [0,0,0]
whole = [0,0,0]
for ele in reader_lines:
    if len(ele[0])> 7 and float(ele[8]) > 0.5:
        if ele[5] == label[0]:
            no_none[0] += float(ele[8])
            whole[0] += float(ele[17])
        elif ele[5] == label[1]:
            no_none[1] += float(ele[8])
            whole[1] += float(ele[17])
        else:
            no_none[2] += float(ele[8])
            whole[2] += float(ele[17])
print(no_none, whole)
print((whole[0] - whole[2]) / whole[1],(whole[0] - whole[1]) / whole[2] )


# distinct value 
```

```python
file = open('./demo_asn.csv','r',encoding='utf-8',newline='')
reader_lines = csv.reader(file)
label = ['total', 'attack', 'normal']
attack_entropy = []
normal_entropy = []
for ele in reader_lines:
    if len(ele[0])> 7 and float(ele[8]) > 0.5:
        if ele[5] == label[1]:
            attack_entropy.append(float(ele[10]))
        if ele[5] == label[2]:
            normal_entropy.append(float(ele[10]))
att_mean = sum(attack_entropy) / len(attack_entropy)
att_std = np.std(attack_entropy)

nor_mean = sum(normal_entropy) / len(normal_entropy)
nor_std = np.std(normal_entropy)
print(att_mean,att_std,nor_mean,nor_std)
```
